### Merging Dataframe with country names data Frame

```python
cd=pd.read_excel("C:\\Users\\Dell\\OneDrive\\Desktop\\excel books\\Country-Code.xlsx")
```

```python
data=pd.merge(df,cd,how='left',on='Country Code')
```

```python
data.info()
```

```python
data.head()
```

```python
countries=data.Country.value_counts().index
print(countries)
```

```python
country_wise=data.Country.value_counts().values
```

```python
plt.pie(country_wise,labels=countries)
```

```python
## top 3 countries
```

```python
plt.pie(country_wise[:3],labels=countries[:3],autopct='%1.2f%%')
```

### observation : Major contribution is from india , followed by United States and then United Kingdom 

```python
data.columns
```

```python
data.groupby(['Aggregate rating','Rating color','Rating text']).size() 
```

```python
review=data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
review=review.rename(columns={0:'Rating_count'}) 
```

```python
plt.figure(figsize=(13,6))
sn.barplot(x='Aggregate rating',y='Rating_count',data=review,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

```python
plt.figure(figsize=(6,6))
sn.countplot(width=1.5,x='Rating color',data=review,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
```
